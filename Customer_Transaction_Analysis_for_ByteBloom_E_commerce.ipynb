## **1. Domain Explanation: The World of E-commerce & Retail**

Before diving into the data, it's crucial to understand the business context. The e-commerce and retail sector is a dynamic and highly competitive industry where data is the key to success.

### **The Business Model: How E-commerce Works**

Modern e-commerce companies are complex operations that go far beyond just having a website. Their business model revolves around a few core pillars:

1.  **Inventory & Supply Chain:** A company must decide how to source its products. Some hold their own inventory in warehouses, giving them control over stock and shipping (like ByteBloom). Others use a **dropshipping** model, where they pass orders directly to a third-party manufacturer who then ships to the customer. The supply chain involves everything from procuring products from suppliers, managing warehouse stock, to logistics for order fulfillment and delivery. An efficient supply chain is critical for ensuring products are available (**in-stock**) and delivered to customers quickly.

2.  **Marketing & Sales:** Unlike a physical store, an e-commerce business must actively drive traffic to its website. This is done through a multi-channel marketing strategy, including **Search Engine Optimization (SEO)** to rank high on Google, **Search Engine Marketing (SEM)** through paid ads, social media campaigns, email marketing, and influencer partnerships. The goal of these efforts is to attract potential customers and guide them through a sales funnel, from awareness to final purchase.

3.  **Customer Service:** In a digital marketplace, trust is paramount. Excellent customer service builds that trust. This includes providing support for product inquiries, managing order issues, and handling returns and refunds smoothly. A positive customer service experience is often the deciding factor in whether a first-time buyer becomes a loyal, repeat customer.

### **The Importance of Data**

In e-commerce, customer transaction data is the **lifeblood** of the business. Every single purchase, click, and interaction creates a data point. This data is not just a record of sales; it's a direct reflection of customer behavior, preferences, and needs. By analyzing this data, a company can move from making decisions based on guesswork to making them based on evidence. Transaction data can reveal which products are popular, who the most valuable customers are, when customers are most likely to buy, and which marketing strategies are actually working. Without analyzing this data, a business is flying blind.

### **Key Performance Indicators (KPIs)**

To measure success and health, e-commerce businesses rely on several critical KPIs:

* **Customer Lifetime Value (CLV):** This metric predicts the total net profit a company can expect to make from a single customer over the entire duration of their relationship. A high CLV indicates strong customer loyalty and profitability. Businesses use CLV to justify spending more to acquire high-value customers.
* **Average Order Value (AOV):** AOV measures the average dollar amount a customer spends per transaction. Increasing AOV is a key goal, often pursued through strategies like product bundling, upselling, and offering free shipping for orders above a certain threshold.
* **Customer Acquisition Cost (CAC):** CAC is the total cost of sales and marketing efforts required to acquire a new customer. It's crucial to ensure that a customer's lifetime value (CLV) is significantly higher than the cost to acquire them (CAC).
* **Churn Rate:** This measures the percentage of customers who stop doing business with a company over a given period. A high churn rate is a major red flag, as it's far more expensive to acquire a new customer than to retain an existing one. Analyzing the behavior of churned customers can reveal pain points in the customer experience.
* **Conversion Rate:** This is the percentage of website visitors who complete a desired action, usually making a purchase. A low conversion rate might indicate issues with the website's design, product pricing, or checkout process.

### **Common Challenges & The Role of Data Analysis**

E-commerce companies face persistent challenges that data analysis can directly address:

* **Challenge: Cart Abandonment.** Many users add items to their cart but leave without completing the purchase.
    * **Data Analysis Solution:** By analyzing where users drop off in the checkout process, companies can identify and fix friction points. They can also use data to trigger automated follow-up emails with reminders or special offers.
* **Challenge: Customer Retention.** Keeping customers coming back is difficult in a competitive market.
    * **Data Analysis Solution:** Analyzing purchasing history allows for the creation of targeted loyalty programs and personalized marketing campaigns that make customers feel valued. **RFM Analysis** (Recency, Frequency, Monetary) is a common technique to segment customers and identify the most loyal ones.
* **Challenge: Inventory Management.** Holding too much stock (overstock) ties up capital, while holding too little (stockout) leads to lost sales.
    * **Data Analysis Solution:** Sales data can be used to forecast future demand for specific products, helping the company optimize inventory levels, reduce waste, and ensure popular items are always available.
* **Challenge: Personalization at Scale.** Customers now expect personalized recommendations and offers.
    * **Data Analysis Solution:** By analyzing browsing history and past purchases, businesses can build sophisticated recommendation engines (like "Customers who bought this also bought...") that significantly improve the user experience and increase AOV.

In summary, data analysis is not just a technical function but a core strategic driver in e-commerce, providing the actionable insights needed to grow revenue, improve efficiency, and build lasting customer relationships.

8.  Parse the `customer_phone` numbers to a standard E.164 format (e.g., +1XXXXXXXXXX). (This is a complex core python/regex task).



```python
print("Unique values in 'customer_phone' before standardization (first 5):")
print(df['customer_phone'].head(5))
```

```python
# Write your code here
import re

def standardize_phone(phone_number):
  if pd.isna(phone_number):
    return None

  # Remove non-digit characters
  digits = re.sub(r'\D', '', str(phone_number))

  # Handle common patterns and convert to E.164
  if len(digits) == 10: # Assuming it is US number
    return '+1' + digits
  elif len(digits) == 11 and digits.startswith('1'):
    return '+' + digits
  elif len(digits) > 11:
    if digits.startswith('00'):
      return '+' + digits[2:]
    if not digits.startswith('+'):
      return '+'+digits
  return str(phone_number)

df['customer_phone'] = df['customer_phone'].apply(standardize_phone)
print("Unique values in 'customer_phone' after standardization (first 5):")
print(df['customer_phone'].head())
```

16. Create a `customer_age` column by calculating the age from `customer_dob`.


```python
# Write your code here
current_date = pd.to_datetime('today')
df['customer_age'] = (current_date.year - df['customer_dob'].dt.year) \
                          - ((current_date.month < df['customer_dob'].dt.month) | \
                             ((current_date.month == df['customer_dob'].dt.month) & (current_date.day < df['customer_dob'].dt.day)))


print("Customer Age created. First 5 values:")
print(df['customer_age'].head())
print(f"Number of missing customer ages: {df['customer_age'].isnull().sum()}")
```

25. Analyze the refund rate (`transaction_status` == 'Refunded') across different `product_category`s. Which categories have the highest refund rates?


**Reasoning**: To analyze the refund rate, I will first filter the transactions that were refunded, then calculate the total transactions and refunded transactions per product category, and finally compute the refund rate and display the results.

```python
refunded_transactions = df[df['transaction_status'] == 'Refunded']

refund_counts = refunded_transactions.groupby('product_category').size()

total_transactions_per_category = df.groupby('product_category').size()

refund_rates = (refund_counts / total_transactions_per_category).fillna(0)

refund_rates = refund_rates.sort_values(ascending=False)

print("Refund Rates by Product Category:")
print(refund_rates)
```
